```python
import requests  # httpx.Session does not exist :(
from bs4 import BeautifulSoup
```

```python
session = requests.Session()
password_reset_url = "http://localhost:8000/accounts/password/reset/"
r = session.get(password_reset_url)
soup = BeautifulSoup(r.text, 'html.parser')
csrf_token = soup.find('input', {'name': 'csrfmiddlewaretoken'}).get('value')
print(csrf_token)
data = {"email": "foo@example.org"}
headers = {
    "X-CSRFToken": csrf_token,
    "Referer": password_reset_url,
}
for i in range(10):
    r = session.post(password_reset_url, data=data, headers=headers)
    print(r.status_code)
    # print(r.content)
```
